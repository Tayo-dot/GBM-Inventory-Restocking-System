```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("/", "_")
)
```

```python
df.columns
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
numeric_cols = [
    'inventory_level', 'units_sold', 'price',
    'discount', 'competitor_pricing'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df = df[
    (df['units_sold'] >= 0) &
    (df['inventory_level'] >= 0) &
    (df['price'] > 0)
]
```

```python
df = df.sort_values(by=['product_id', 'date']).reset_index(drop=True)
```

```python
from sklearn.preprocessing import LabelEncoder

label_cols = [
    'store_id', 'product_id', 'category',
    'region', 'weather_condition', 'seasonality'
]

for col in label_cols:
    if col in df.columns:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
```

```python
df_feat = df.sort_values(['store_id','product_id','date']).reset_index(drop=True)

df_feat['lag_1'] = df_feat.groupby(['store_id','product_id'])['units_sold'].shift(1)
df_feat['lag_7'] = df_feat.groupby(['store_id','product_id'])['units_sold'].shift(7)

df_feat['roll_mean_7'] = (
    df_feat.groupby(['store_id','product_id'])['units_sold']
    .shift(1).rolling(7).mean()
)

df_feat['roll_mean_14'] = (
    df_feat.groupby(['store_id','product_id'])['units_sold']
    .shift(1).rolling(14).mean()
)

df_feat['roll_std_7'] = (
    df_feat.groupby(['store_id','product_id'])['units_sold']
    .shift(1).rolling(7).std()
)

# Time features
df_feat['month'] = df_feat['date'].dt.month
df_feat['day_of_week'] = df_feat['date'].dt.dayofweek
df_feat['day'] = df_feat['date'].dt.day

# Remove rows that don't have enough history (because lag/rolling will create NaNs at the start)
df_feat = df_feat.dropna().reset_index(drop=True)
print("After lag features, rows:", len(df_feat))
```

```python
df_feat.columns
```

```python
drop_cols = ['date']
X = df_feat.drop(columns=drop_cols + ['units_sold'], errors='ignore')
y = df_feat['units_sold']
```

```python
split_index = int(len(df_feat) * 0.8)

X_train = X.iloc[:split_index]
X_test  = X.iloc[split_index:]

y_train = y.iloc[:split_index]
y_test  = y.iloc[split_index:]
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(
    n_estimators=600,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42
)

model.fit(X_train, y_train)
print("Model trained ")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_pred = model.predict(X_test)

print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2:", r2_score(y_test, y_pred))
```

```python
X_no_inventory = X.drop(columns=['inventory_level'])

split_index = int(len(X_no_inventory) * 0.8)

X_train = X_no_inventory.iloc[:split_index]
X_test  = X_no_inventory.iloc[split_index:]
y_train = y.iloc[:split_index]
y_test  = y.iloc[split_index:]

_ = model.fit(X_train, y_train)

y_pred = model.predict(X_test)

from sklearn.metrics import r2_score
print("R2:", r2_score(y_test, y_pred))
```

```python
df_feat['lag_3'] = df_feat.groupby(['store_id','product_id'])['units_sold'].shift(3)
df_feat['lag_14'] = df_feat.groupby(['store_id','product_id'])['units_sold'].shift(14)

df_feat['roll_mean_30'] = (
    df_feat.groupby(['store_id','product_id'])['units_sold']
    .shift(1).rolling(30).mean()
)

df_feat['roll_std_14'] = (
    df_feat.groupby(['store_id','product_id'])['units_sold']
    .shift(1).rolling(14).std()
)

df_feat = df_feat.dropna().reset_index(drop=True)
```

```python
X = df_feat.drop(columns=['units_sold','date'])
y = df_feat['units_sold']

split_index = int(len(df_feat) * 0.8)

X_train = X.iloc[:split_index]
X_test  = X.iloc[split_index:]
y_train = y.iloc[:split_index]
y_test  = y.iloc[split_index:]
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(
    n_estimators=2000,
    learning_rate=0.02,
    max_depth=10,
    min_child_weight=3,
    subsample=0.8,
    colsample_bytree=0.8,
    gamma=0.1,
    reg_alpha=1,
    reg_lambda=2,
    random_state=42,
    early_stopping_rounds=100
)
_ = model.fit(
    X_train,
    y_train,
    eval_set=[(X_test, y_test)],
    verbose=False
)
print("Model trained with early stopping ")
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

y_pred = model.predict(X_test)

print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2:", r2_score(y_test, y_pred))
```
